Scrapping 

```python
#HEADERS = ({'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})

HEADERS = [({'User-Agent':
            'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'}),({'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})
          ]

def details_part(soup):
    
    s=str(soup.find("div",attrs={"id": 'detailBullets_feature_div'}).get_text())
    s=s.replace("\n\n"," ")
    s=s.replace('\u200f',"\n")
    s=s.replace('\u200e',"\n")
    s=s.replace("  ","")
    s=s.replace("\n","")
    return s

c=0
D={}
for i in range(len(df)):

    try:
        url="https://www.amazon."+df['country'][i]+"/dp/"+df['Asin'][i]
        webpage = requests.get(url, headers=HEADERS[0])
        soup = BeautifulSoup(webpage.content, "lxml")
        sleep(4)
        d={}
        
        prod_title=str(soup.find("span",attrs={"id": 'productTitle'}).get_text())
        prod_img=soup.find("div",attrs={"id": "img-canvas"}).img['src'] ########3
        prod_price=str(soup.find("span",attrs={"id": 'price'}).text) #########
        prod_det=details_part(soup)   ##############33333
        d['Product Title']=prod_title
        d['Product Image URL']=prod_img
        d['Price of the Product']=prod_det
        D[c]=d
        c+=1
    except:
        try:
            url="https://www.amazon."+df['country'][i]+"/dp/"+df['Asin'][i]
            webpage = requests.get(url, headers=HEADERS[1])
            soup = BeautifulSoup(webpage.content, "lxml")
            sleep(4)
            d={}

            prod_title=str(soup.find("span",attrs={"id": 'productTitle'}).get_text())
            prod_img=soup.find("div",attrs={"id": "img-canvas"}).img['src'] ########3
            prod_price=str(soup.find("span",attrs={"id": 'price'}).text) #########
            prod_det=details_part(soup)   ##############33333
            d['Product Title']=prod_title
            d['Product Image URL']=prod_img
            d['Price of the Product']=prod_det
            D[c]=d
            c+=1
        except:
            print(url, " --> Not Available")
```

Json 

```python
with open("output.json", "w") as outfile:
    json.dump(D, outfile)
D
```

creating a table from the json

```python
df1=pd.read_json('output.json', orient='index')
```

Upload to SQL

```python
engine = create_engine('sqlite://', echo=False)
```

```python
df1.to_sql('users', con=engine)
engine.execute("SELECT * FROM users").fetchall()
```
